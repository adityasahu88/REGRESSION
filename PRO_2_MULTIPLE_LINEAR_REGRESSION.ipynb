### Visualization on Avg. Session Length	,Time on App,	Time on Website	Length of Membership,	Yearly Amount Spent

```python
fig,axes = plt.subplots(2,2,figsize=(20,10))
axes[0,0].plot(df['Avg. Session Length'],df['Yearly Amount Spent'],'mo')
axes[0,0].set_ylabel("Yearly Amount Spent", fontsize = 10)
axes[0,0].set_xlabel("Avg. Session Length", fontsize = 10)

axes[0,1].plot(df['Time on App'],df['Yearly Amount Spent'],'o')
axes[0,1].set_ylabel("Yearly Amount Spent", fontsize = 10)
axes[0,1].set_xlabel("Time on App", fontsize = 10)

axes[1,0].plot(df['Time on Website'],df['Yearly Amount Spent'],'go')
axes[1,0].set_ylabel("Yearly Amount Spent", fontsize = 10)
axes[1,0].set_xlabel("Time on Website", fontsize = 10)

axes[1,1].plot(df['Length of Membership'],df['Yearly Amount Spent'],'ro')
axes[1,1].set_ylabel("Yearly Amount Spent", fontsize = 10)
axes[1,1].set_xlabel("Length of Membership", fontsize = 10)
```

# Splitting data into test and train

```python
X_train,X_test,y_train,y_test =train_test_split(df[["Avg. Session Length","Time on App","Time on Website","Length of Membership"]],
                                                df[["Yearly Amount Spent"]],random_state=42,test_size=0.20)
```

# Applying Linear Regression

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
print("Intercept:", lr.intercept_.round(2))
print("Coefficient:", lr.coef_.round(2))
```

```python
y_pred = lr.predict(X_test)
```

# Evaluating the model

```python

print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('MAE:', mean_absolute_error(y_test, y_pred))
print('R2:', r2_score(y_test, y_pred))
```
